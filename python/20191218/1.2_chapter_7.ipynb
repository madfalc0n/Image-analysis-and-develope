### 7.2.5 Discretization and Binning

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
bins = [18, 25, 35, 60, 100]
```

```python
#cats = pd.cut(ages, bins) #연속데이터를 개별로 분할하거나 분석을위해 그룹으로 나누고 싶을 때 cut 함수 사용!
#cats
cats = pd.cut(ages,bins) 
print(cats)
print(type(cats)) #카테고리컬 타입
```

```python
print(cats.codes)
print(cats.categories)
print(pd.value_counts(cats)) #pandas.cut 결과에 대한 그룹 수를 나타냄
```

```python
pd.cut(ages, [18, 26, 36, 61, 100], right=False)
```

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_names)
```

```python
data = np.random.rand(20)
print(data)
pd.cut(data, 4, precision=2) #균등하게 4개의 그룹으로 나뉨, precision 이 2이면 소수점 2자리까지 제한
```

```python
data = np.random.randn(1000)  # Normally distributed, 정규분포를 따르는 난수 생성
print(data)
cats = pd.qcut(data, 4)  # Cut into quartiles , 표본 변위치를 기반으로 데이터를 나눠줌
cats
pd.value_counts(cats)
```

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]) # 변위치를 직접 지정도 가능
```

### 7.2.6 Detecting and Filtering Outliers(특이값 찾고 제외하기)

```python
data = pd.DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
col = data[2]
col[np.abs(col) > 3]
```

```python
data[(np.abs(data) > 3).any(1)] # 각 열에서 값이 3인 애들이 하나라도 있으면 출력
```

```python
data[np.abs(data) > 3] = np.sign(data) * 3 # 3초과 하는 애들을 -3 또는 3으로 지정
data.describe()
```

#### 7.2.8 Computing Indicator/Dummy Variables(표시자/더미변수 계산하기)

```python
df = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                   'data1': range(6)})
pd.get_dummies(df['key']) #one-hot 인코딩 방식과 같다고 보면 된다.
```

```python
dummies = pd.get_dummies(df['key'], prefix='keys') #prefix 는 더미 테이블 생성될 때 column 양식
df_with_dummy = df[['data1']].join(dummies)
df_with_dummy
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('datasets/movielens/movies.dat', sep='::',
                       header=None, names=mnames)
movies[:10]
```

```python
all_genres = []
for x in movies.genres:
    all_genres.extend(x.split('|'))
genres = pd.unique(all_genres)
```

```python
genres
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
```

```python
gen = movies.genres[0]
gen.split('|')
dummies.columns.get_indexer(gen.split('|'))
```

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[0]
```
