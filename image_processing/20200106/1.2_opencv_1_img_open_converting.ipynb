```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

### 이미지 불러오기

```python
img = cv2.imread('./data/fish.jpg') #넘파이 타입으로 리턴한다.
print(type(img))
print(img.shape) #영상 크기 반환
plt.imshow(img) #영상표시, 색이 이상하게 나온다.채널의 순서가 다르기 때문, opencv는 RGB 채널이 아닌 BGR 순이다
```

## 이미지 색 변경 

```python
b = img[:, :, 0] # 블루채널, 첫행과 컬럼은 전부 가져오고,depth는 0번째 만 
print(b)
g = img[:, :, 1] #
print(g)
r = img[:, :, 2] #
print(r)

sum_image = np.hstack([b, g, r])
plt.imshow(sum_image)
plt.show()

rgb = np.dstack([r,g,b]) #원래 이미지 복원
plt.imshow(rgb)
plt.show()
```

```python
b = img[:, :, 0] + 2
print(b.shape)
g = img[:, :, 1] + 2
print(g.shape)
r = img[:, :, 2] + 2
print(r.shape)

sum_image = np.hstack([b, g, r])
plt.imshow(sum_image)
plt.show()

rgb = np.dstack([r,g,b]) #원래 이미지 복원
plt.imshow(rgb)
plt.show()
```

## color converting 함수

```python
#BGR -> RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
cv2.imwrite('./data/result.jpg',rgb) #저장할때는 BGR로 저장됨
```
