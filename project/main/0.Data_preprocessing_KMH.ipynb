```python
import pandas as pd
import numpy as np
```

### 1. 서울 자치구 별 범죄수 데이터 전처리

```python
crime_df = pd.read_csv('./data/crime_list/seoul_crime_len(2014~2018).txt', sep='\t', encoding='UTF8', header=None)
crime_df.head(5)
# 값이 '검거'인 항목을 제거하고 1번 컬럼 삭제
occur_crime = crime_df[1] != '검거'
crime_df = crime_df[occur_crime]
crime_df.drop([1], axis=1, inplace=True)
#display(crime_list.shape)

#전치행렬 변환 및 컬럼 재정의
t_crime_df = crime_df.T
t_crime_df.columns = ['년도', '지역', '범죄합계', '살인', '강도', '강간강제추행', '절도', '폭력']
t_crime_df.drop([0],axis=0, inplace = True)
t_crime_df.reset_index(drop=True, inplace=True)
print(t_crime_df.info())

#데이터 형변환
t_crime_df[['년도','범죄합계', '살인', '강도', '강간강제추행', '절도', '폭력']] = t_crime_df[['년도','범죄합계', '살인', '강도', '강간강제추행', '절도', '폭력']].astype(int)
t_crime_df

#지역 항목중 '합계'인 항목 삭제
place_sum = t_crime_df['지역'] != '합계'
t_crime_df = t_crime_df[place_sum]

#피벗 테이블 생성
pivot_t_crime_df = t_crime_df.pivot_table(index=['년도','지역'])
display(pivot_t_crime_df)
```

### 2. 서울 자치구별 CCTV 수 전처리

```python
#데이터셋 호출
cctv_df = pd.read_excel('./data/cctv_list/seoul_CCTV_2014~2018.xlsx')

#'지역' 컬럼 데이터 내 공백 제거
cctv_df['기관명'] = cctv_df['기관명'].str.replace(' ','')

# 불필요 항목 제거('소계')
cctv_df.drop(['소계'], axis=1,inplace=True)
display(cctv_df)

# 인덱스 이름 변경 및 컬럼 재정의
t_cctv_df = cctv_df.T
t_cctv_df.rename(index={'기관명':'년도'}, inplace=True)
t_cctv_df.columns = t_cctv_df.loc['년도']
t_cctv_df.drop(['년도'], axis=0, inplace=True)
display(t_cctv_df)

#범죄데이터와 CCTV 수 병합
for i in t_cctv_df.index:
    for j in t_cctv_df.columns:
        pivot_t_crime_df.loc[(i,j ), 'CCTV'] = int(t_cctv_df.loc[i, j])    

display(pivot_t_crime_df)
```

### 3. 서울 자치구별 인구수 데이터 전처리

```python
#결측치 처리 및 필요없는 컬럼 지우기
popul_df = pd.read_excel('./data/pop_list/seoul_population_(2014~2018).xls')
display(popul_df.shape)

#'기간' 컬럼 데이터에서 결측치 처리 및 '지역'컬럼 데이터에서 '함계' 값 제외
popul_df['기간'].fillna(method='ffill', inplace=True)
popul_df = popul_df[popul_df['지역'] != '합계']

#형변환
popul_df['기간'] = popul_df['기간'].astype(int)
popul_df.info()

#피벗테이블로 변환
pivot_popul_df = popul_df.pivot_table(index=['기간','지역']).astype(int) #피벗테이블은 기본 평균으로 병합됨
display(pivot_popul_df.info())

#합계 컬럼 rename
pivot_popul_df.rename(columns = {'합계': '인구합계'}, inplace = True)

#기존 테이블과 병합 및 컬럼 형변환
display(pivot_popul_df)
```

### 4. 서울 자치구별 식품위생업 데이터 전처리

```python
#base 데이터셋 호출
food_df = pd.read_table('./data/food_list/서울시 식품위생업 현황 (구별) 통계(2014).txt', header=None)
# 데이터 호출 및 concat을 통한 병합
year_list = ['2015', '2016', '2017', '2018']
for i in year_list:
    temp_df = pd.read_table('./data/food_list/서울시 식품위생업 현황 (구별) 통계('+ i +').txt', header=None)
    temp_df.drop([0,1],axis=1 ,inplace=True)
    food_df = pd.concat([food_df,temp_df],axis=1)
#특정데이터만 추출
imp_col = (food_df[1] == '단란주점') | (food_df[1] == '유흥주점') | (food_df[1] == '구분(2)')
food_df = food_df[imp_col]
#데이터프레임 전치
t_food_df = food_df.T
#필요없는 행 삭제
t_food_df.drop([0,1,2],axis=0 ,inplace=True)
#컬럼 지정
t_food_df.columns = ['년도','지역','단란주점','유흥주점']
#인덱스 재조정
t_food_df.reset_index(drop=True,inplace=True)
#형변환
#display(food_df.head(3))
t_food_df[['년도','단란주점','유흥주점']] = t_food_df[['년도','단란주점','유흥주점']].astype(int)
t_food_df.info()
#피벗테이블로 변환
pivot_t_food_df = t_food_df.pivot_table(index=['년도','지역'])
#단란+유흥
pivot_t_food_df['단란유흥'] =  pivot_t_food_df['단란주점'] + pivot_t_food_df['유흥주점']
display(pivot_t_food_df)
#pivot_t_food_df.to_csv('./data/20191231_seoul_식품위생업_데이터_2014_to_2018.csv')
```

###  5. 모든데이터 병합

```python
# 범죄+CCTV , 인구수 , 식품위생업 병합
seoul_hangjeong_data_2014_to_2018 = pd.concat([pivot_t_crime_df, pivot_popul_df,pivot_t_food_df] , axis =1)

#10만인구별 범죄 및 CCTV 수 계산
pup_10man = 100000 / seoul_hangjeong_data_2014_to_2018['인구합계']
seoul_hangjeong_data_2014_to_2018['10만명당CCTV수'] = seoul_hangjeong_data_2014_to_2018['CCTV']*pup_10man
seoul_hangjeong_data_2014_to_2018['10만명당범죄수'] = seoul_hangjeong_data_2014_to_2018['범죄합계']*pup_10man

# 컬럼 형변환 
seoul_hangjeong_data_2014_to_2018[seoul_hangjeong_data_2014_to_2018.columns] = seoul_hangjeong_data_2014_to_2018[seoul_hangjeong_data_2014_to_2018.columns].astype(int)
display(seoul_hangjeong_data_2014_to_2018)

#to_csv
seoul_hangjeong_data_2014_to_2018.to_csv('./data/20200101_seoul_2014_to_2018_crime_cctv_pop_food_sum.csv')
```
