```python
import pandas as pd
import folium

file_path='./data/2016_middle_shcool_graduates_report.xlsx'
df =pd.read_excel(file_path, header=0)

print(df.info())
print(df.columns.values)
```

```python
#참고 url : http://python-graph-gallery.com/288-map-background-with-folium
mschool_map = folium.Map(location=[37.55, 126.98], tiles='Stamen Terrain', zoom_start=12)
for name, lat, lng in zip (df.학교명, df.위도, df. 경도) :
    folium.CircleMarker([lat, lng],
                              radius=5,
                              color='brown',
                              fill=True,
                              fill_color='yellow',
                              fill_opacity=0.5,
                              popup=name).add_to(mschool_map)

mschool_map.save('./data/seoul_mschool_location.html')
```

```python
#지역, 코드, 유형, 주야 등 변수의 one-hot-encoding으로 dummy 변수 변환
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()  #한 컬럼에 데이터 별로 다른 숫자가 들어간다.
onehot_encoder = preprocessing.OneHotEncoder()
```

```python

onehot_location = label_encoder.fit_transform(df['지역'])
print(df['지역'])
print(df['지역'].unique())
print(onehot_location)
print()

onehot_code = label_encoder.fit_transform(df['코드'])
print(df['코드'])
print(df['코드'].unique())
print(onehot_code)
print()


onehot_type = label_encoder.fit_transform(df['유형'])
print(df['유형'])
print(df['유형'].unique())
print(onehot_type)
print()

onehot_day= label_encoder.fit_transform(df['주야'])
print(df['주야'])
print(df['주야'].unique())
print(onehot_day)
print()
```

```python
df['location'] = onehot_location
df['code'] = onehot_code
df['type'] = onehot_type
df['day'] = onehot_day

print(df.head())
print(df.info())
```

```python
#DBSCAN 군집분석 수행
from sklearn import cluster
#분석에 사용할 속성 [과학고, 외고_국제고, 자사고 ] 진학률
column_list = [9, 10, 13]
X = df.iloc[ : , column_list]
print(X[:5])

```

```python
# 설명변수 정규화
X= preprocessing.StandardScaler().fit(X).transform(X)
print(X)
dbscan_model = cluster.DBSCAN(eps=0.2, min_samples=5)
dbscan_model.fit(X)

cluster_label = dbscan_model.labels_
print(cluster_label)

df['Cluster'] = cluster_label
print(df.head())
print('\n')
```

```python
#클러스터값으로  그룹화하여 출력
grouped_cols =[0, 1, 3] +column_list
grouped = df.groupby('Cluster')
for key, group in grouped:
    print('* key : ' , key)
    print('* number : ' , len(group))
    print(group.iloc[:, grouped_cols].head())
    print('\n')

colors = {-1:'gray', 0:'coral', 1:'black', 2:'green', 3:'red', 4:'purple', 5:'orange', 6:'brown', 7:'magenta', 8:'yellow', 9:'brick', 10:'cyan', 11:'pink'}
```

```python
DBSCAN_cluster_school_map = folium.Map(location=[37.55, 126.98], tiles='Stamen Terrain', zoom_start=12)
for name, lat, lng, clus in zip (df.학교명, df.위도, df. 경도, df.Cluster) :
    folium.CircleMarker([lat, lng],
                              radius=5,
                              color=colors[clus],
                              fill=True,
                              fill_color=colors[clus] ,
                              fill_opacity=0.5,
                              popup=name).add_to(DBSCAN_cluster_school_map)
```
