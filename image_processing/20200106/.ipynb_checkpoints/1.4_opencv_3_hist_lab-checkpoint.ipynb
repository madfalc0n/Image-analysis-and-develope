```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
train = cv2.imread('./data/patch.png')
print(train.shape)

#hstack을 이용해 옆으로 BGR 순으로 그림표시
plt.imshow(np.hstack([b,g,r]))
plt.show()
```

## Histogram
1. BGR로 표시

```python
b,g,r = cv2.split(train) # 3개의 채널을 분리해줌
#print(b.shape,g.shape,r.shape)(172, 154) (172, 154) (172, 154)
# print(b.ravel().shape) # 172 * 154 = 26488 # (26488,)
# print(b.shape) #(172, 154)

#ravel 함수는 2차원행렬을 1차원 벡터로 바꿔주는 함수, bin - 256개 , 0~ 256 사이
plt.hist(b.ravel(), 256, [0,256], color='b') # 블루의 분포도 , 130 ~ 200 사이
plt.show()
plt.hist(g.ravel(), 256, [0,256], color='g' ) # 그린의 분포도, 130 ~ 200 사이
plt.show()
plt.hist(r.ravel(), 256, [0,256], color='r' ) # 레드의 분포도, 100 ~ 200 사이
plt.show()

plt.hist(b.ravel(), 256, [0,256] , color = 'b') #ravel 함수는 2차원행렬을 1차원 벡터로 바꿔주는 함수, bin - 256개 , 0~ 256 사이
plt.hist(g.ravel(), 256, [0,256] , color = 'g') 
plt.hist(r.ravel(), 256, [0,256] , color = 'r')  
plt.show() #생성마다 쓰면 따로 그림
```

## Histogram
1. HSV로 표시

```python
h,s,v = cv2.split(cv2.cvtColor(train, cv2.COLOR_BGR2HSV)) # 3개의 채널을 분리해줌
print(h.shape,s.shape,v.shape)
plt.imshow(np.hstack([h,s,v]))
plt.show()

plt.hist(h.ravel(), 256, [0,256] , color = 'b') # h 색종류 빨~주~노~초~ , 50에서 100 사이의 값(색)의 각각 value_counts
plt.hist(s.ravel(), 256, [0,256] , color = 'g') # s 채도 흰색에서 초록 , 0에서 50 사이에 값(선명도)의 각각 value_counts
plt.hist(v.ravel(), 256, [0,256] , color = 'r') # v 명도 어두운영역부터 밝은 영역 , 약 130에서 200 사이에 값(밝기)의 각각 value_counts
plt.show() #생성마다 쓰면 따로 그림
```
