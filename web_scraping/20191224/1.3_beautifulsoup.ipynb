```python
from bs4 import BeautifulSoup
# 분석하고 싶은 HTML  루트 요소인 <html>에서 마침표(.)를 사용해 값에 접근
html = """
<html><body>
  <h1>스크레이핑이란?</h1>
  <p>웹 페이지를 분석하는 것</p>
  <p>원하는 부분을 추출하는 것</p>
</body></html>
"""

html2 = "www.naver.com"
# HTML 분석하기   
soup = BeautifulSoup(html, 'html.parser')
# 원하는 부분 추출하기  
h1 = soup.html.body.h1
print(h1)
p1 = soup.html.body.p
print(p1)
p2 = p1.next_sibling.next_sibling
print(p2)
# 요소의 글자 출력하기  
print(type(soup))
print("h1 = " + h1.string)
print("p1  = " + p1.string)
print("p2  = " + p2.string)

```

```python
from bs4 import BeautifulSoup

html = """
<html><body>
  <h1 id="title">스크레이핑이란?</h1>
  <p id="body">웹 페이지를 분석하는 것</p>
  <p>원하는 부분을 추출하는 것</p>
</body></html>
"""

# HTML 분석하기 
soup = BeautifulSoup(html, 'html.parser')

# find() 메서드로 원하는 부분 추출

title = soup.find(id="title")
body = soup.find(id="body")

# 텍스트 부분 출력하기
print("#title=" + title.string)
print("#body=" + body.string)
```

```python
from bs4 import BeautifulSoup 
html = """
<html><body>
  <ul>
  <h1 id="title">스크레이핑이란?</h1>
  <p>원하는 부분을 추출하는 것</p>
  <p>원하는 부분을 추출하는 것2</p>
    <li><a href="http://www.naver.com">naver</a></li>
    <li><a href="http://www.daum.net">daum</a></li>
  </ul>
</body></html>
"""
# 태그의 속성은 attrs 속성에서 추출
# 내부의 설명 텍스트는 string 속성으로 추출
# HTML 분석하기  
soup = BeautifulSoup(html, 'html.parser')

# find_all() 메서드로 추출, 여러 개의 태그를 한 번에 추출
links2 = soup.find_all("p")
print(links2) #[<p>원하는 부분을 추출하는 것</p>, <p>원하는 부분을 추출하는 것2</p>]
# 링크 목록 출력하기  
links = soup.find_all("a")
for a in links:
    href = a.attrs['href']
    text = a.string
    print(text, ">", href)
```
