```python
import csv
```

```python
f = open("csvfile.csv","r")
reader = csv.reader(
    f,
    delimiter=',',
    quotechar="",
    quoting=csv.QUOTE_ALL)
```

```python
seoung_nam_data = []
header = []
rownum = 0

with open("korea_floating_population_data.csv","r",encoding="cp949") as p_file:
    csv_data = csv.reader(p_file)
    
    for row in csv_data:
        if rownum == 0:
            header =row
        location = row[7]

        if location.find("성남시") != -1:
            print(location)
            seoung_nam_data.append(row)
        rownum +=1
    
    
with open("seoung_nam_floating_population_data.csv", "w",encoding="cp949") as s_p_file:
    writer = csv.writer(s_p_file, delimiter='\t', quotechar="'", quoting=csv.QUOTE_ALL)
    writer.writerow(header)
    
    for row in seoung_nam_data:
        writer.writerow(row)
        
```
