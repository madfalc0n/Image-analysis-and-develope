```python
# 4.2017년 수원시 인구의 합
suwon_pop_sum = 0
for i in range(1,len(data)):
    str_va = str(data[i])
    if str_va.startswith('수원시'):
        #print(data[i,-1])
        suwon_pop_sum += int(data[i,-1])
print(suwon_pop_sum)
```
