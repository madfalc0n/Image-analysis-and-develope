```python
#모듈 호출
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
import re
```

```python
#셀레니움 모듈 호출 및 URL get
driver = webdriver.Chrome('C:/chromedriver/chromedriver.exe')
driver.implicitly_wait(3)
driver.get('http://www.imdb.com/search/title?count=100&release_date=2017,2017&title_type=feature')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# title
title = soup.select('h3 > a')
title_list = []
for i in title:
    temp_str = i.text
    title_list.append(temp_str)
movie_dict['title'] = title_list
title_list
```

```python
# description
description = soup.select('.lister-item-content > .text-muted:nth-of-type(2)')
description_list = []
for i in description:
    temp_str = i.text
    temp_str.strip()
    temp_str = re.sub('[\n]','',temp_str)
    #print(temp_str)
    description_list.append(temp_str)
movie_dict['description'] = description_list
description_list
```

```python
# runtime, 완
runtime = soup.select('p > .runtime')
runtime_list = []
for i in runtime:
    temp_str = i.text
    runtime_list.append(temp_str)
movie_dict['runtime'] = runtime_list
```

```python

# genre, 완
genre = soup.select('p > .genre')
genre_list = []
for i in genre:
    temp_str = i.text
    genre_list.append(temp_str)
movie_dict['genre'] = genre_list
```

```python

# rating
rating = soup.select('.lister-item-content > .ratings-bar')
rating_list = []
for i in rating:
    
    temp_str = i.text
    print(temp_str)
    rating_list.append(temp_str)
movie_dict['rating'] = title_list
```

```python
# metascore
metascore = soup.select('.inline-block.ratings-metascore')
metascore_list = []
for i in metascore:
    print(i)
    temp_str = i.text
    metascore_list.append(temp_str)
movie_dict['metascore'] = metascore_list
#metascore_list
```
