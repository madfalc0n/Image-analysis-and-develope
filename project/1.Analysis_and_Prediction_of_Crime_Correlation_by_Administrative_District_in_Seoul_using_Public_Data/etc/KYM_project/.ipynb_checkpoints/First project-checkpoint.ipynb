```python
import numpy as np
import pandas as pd
```

```python
crime = pd.read_excel('./data/Report.xls')

crime.drop(['합계.1','살인.1','강도.1','강간강제추행.1','절도.1','폭력.1'],axis=1,inplace=True)

#crime.set_index('기간',inplace=True)

crime = crime[~crime['자치구'].str.contains('합계')]
# display(crime)
#crime.drop('기간',inplace=True)
# display(crime)
crime.drop([0],axis=0,inplace=True)
display(crime)
crime['기간']=crime['기간'].astype(int)

crime = crime.groupby(['기간','자치구']).sum()

# crime['기간'].astype(int)
display(crime)

# crime=crime.copy()
# crime1 = crime.reset_index(drop=True, inplace=True)
```

```python
cctv = pd.read_excel('./data/서울시 자치구 년도별 CCTV 설치 현황(2011년 이전_2018년).xlsx')

cctv = cctv.drop(['소계','2011년 이전','2012년','2013년'],axis=1)
display(cctv.head())
cctv.columns = ['기관명',2014,2015,2016,2017,2018]
cctv['기관명']=cctv['기관명'].str.replace(" ","")
cctv = cctv.set_index('기관명')

cctv = cctv.unstack(level=0)
display(cctv)
# cctv=cctv.copy()
#cctv.reset_index(drop=True, inplace=True)
```

```python
pop = pd.read_excel('./data/서울시 주민등록인구 (구별) 통계(2014년~2018년).xls',header=1)
pop.set_index('기간',inplace=True)
pop.drop(['합계.1','합계.2','한국인.1','한국인.2','등록외국인.1','등록외국인.2','인구밀도','인구밀도.1','세대당인구','65세이상고령자'],axis=1,inplace=True)
# pop.set_index('기간',inplace=True)
pop = pop[~pop['자치구'].str.contains('합계')]
pop = pop.drop(pop.index[0])
# pop = pop.pivot('기간','자치구')

pop = pop.groupby(['기간','자치구']).sum()
display(pop)
# print(pop)
# pop=pop.copy()
# pop1 = pop.reset_index(drop=True, inplace=True)

# display(pop1)
# print(pop.info())

```

```python
merged = pd.concat([cctv,pop,crime],axis=1)
wtf=merged.to_excel('./data/merged1.xlsx')
```
