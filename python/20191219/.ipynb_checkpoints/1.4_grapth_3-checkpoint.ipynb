```python
import folium
import pandas as pd
import numpy as np
```

```python
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12) #그냥지도

seoul_map.save('./data/seoul.html') 
```

```python
seoul_map1 = folium.Map(location=[37.55, 126.98], tiles = 'Stamen Terrain' , zoom_start=12) #등고선 기준
seoul_map2 = folium.Map(location=[37.55, 126.98], tiles = 'Stamen Toner' , zoom_start=12) #흑백

seoul_map2.save('./data/seoul2.html')
seoul_map1.save('./data/seoul1.html')
```

```python
df = pd.read_excel('./data/서울지역 대학교 위치.xlsx')
df.columns=['대학', '위도', '경도']
df
```

```python
#맵 지정
seoul_map = folium.Map(location=[37.55, 126.98], tiles = 'Stamen Terrain', zoom_start = 13 ) #zoom은 값이 커지면 확대됨
for name, at, lng in zip(df.대학, df.위도, df.경도):
    folium.Marker([at, lng], popup=name).add_to(seoul_map) #Document에 다양한 종류의 마커가 있다
    
seoul_map.save('./data/seoul_colleges.html')
```
