```python
img2 = cv2.imread('./data/Han.jpg')
plt.imshow(img2)
plt.show()
hsv = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV) # BGR to HSV
h = hsv[:, : , 0]
s = hsv[:, : , 1]
v = hsv[:, : , 2]
result = img2.copy()

for r in range(h.shape[0]): #  
    for c in range(h.shape[1]):
        col = ( int(img2[r,c,0]) + int(img2[r,c,1]) + int(img2[r,c,2])) / 3 # 8 + 8 + 8
        result[r,c,0] = col #3
        result[r,c,1] = col
        result[r,c,2] = col

plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))  #     , 8  256  ,  type strong  
plt.show()
```
